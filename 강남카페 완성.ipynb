```python
import folium

gangnam_station = (37.498085, 127.027676)
map = folium.Map(location=gangnam_station, zoom_start=16)

# 중복 위치 처리를 위한 델타 값 설정
delta = 0.0001

# 이미 표시한 위치를 저장하기 위한 set
added_locations = set()

for i in range(len(data)):
    latitude = data.loc[i]['위도']
    longitude = data.loc[i]['경도']
    
    # 중복된 위치인지 확인하고 중복된 위치라면 델타 값을 적용하여 조정
    while (latitude, longitude) in added_locations:
        latitude += delta
        longitude += delta
        
    added_locations.add((latitude, longitude))
    
    marker = folium.Marker([latitude, longitude], icon=folium.Icon(color='red'))
    marker.add_to(map)

marker = folium.Circle(gangnam_station, radius=200, color='skyblue', popup='Cafe', fill_color='skyblue')
marker.add_to(map)

map
```
