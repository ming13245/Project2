```python
import folium
import pandas as pd
from geopy.distance import geodesic # 두 지점 간의 거리 계산

jamsil = folium.Map(location = [37.5131, 127.1006], zoom_start = 10) # zoom_start : 지도 초기 확대 수준
data = pd.read_csv("소상공인시장진흥공단_상가(상권)정보_서울_202312.csv")

# 잠실역 중심 지도
map = folium.Map(location = [37.5131, 127.1006], zoom_start = 10)

# 원을 그릴 위치의 위도와 경도를 설정
circle_lat_lon = [37.5131, 127.1006]

# folium.Circle() 메소드로 원 생성
folium.Circle(
    location = circle_lat_lon,
    radius = 200,  # 원의 반지름 (미터 단위)
    color = 'crimson',
    fill = True, # 테두리
    fill_color = 'crimson'
).add_to(map)

# 데이터프레임에서 원 안에 있는 위치 데이터를 추출합니다
locations_inside_circle = []
for index, row in data.iterrows():
    location = (row['위도'], row['경도'])
    if geodesic(location, circle_lat_lon).meters <= 200:  # 10000미터 이내에 있는지 확인
        locations_inside_circle.append(row['상가업소번호'])

print("원 안에 있는 장소:", locations_inside_circle)
```

```python
print(data["상가업소번호"])
```

```python
# 원 안에 있는 상가 업소 데이터 가져오기
selected_data = data[data["상가업소번호"].isin(locations_inside_circle)]
selected_cafe_data = selected_data[(selected_data["상권업종소분류명"].isin(["카페"]))]
# selected_cafe_data
print(len(selected_cafe_data))
selected_cafe_data.to_csv('jamsilStation_cafe.csv')
```

```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# 잠실역 중심 지도
map = folium.Map(location=[37.5131, 127.1006], zoom_start=10)

# 원을 그릴 위치의 위도와 경도를 설정
circle_lat_lon = [37.5131, 127.1006]

# folium.Circle() 메소드로 원 생성
folium.Circle(
    location=circle_lat_lon,
    radius=200,  # 원의 반지름 (미터 단위)
    color='crimson',
    fill=True,  # 테두리
    fill_color='crimson'
).add_to(map)

data = pd.read_csv("jamsilStation_cafe.csv")

# MarkerCluster 객체 생성
marker_cluster = MarkerCluster().add_to(map)

# 데이터프레임에서 상가 업소의 위치를 지도에 표시
for index, row in data.iterrows():
    location = (row['위도'], row['경도'])  # 상가 업소의 위도와 경도
    popup_text = f"<b>{row['상호명']}</b>"
    folium.Marker(location, popup = popup_text, icon = folium.Icon(icon = 'coffee', prefix = 'fa', color = 'red', icon_size = 35)).add_to(marker_cluster)  # 마커를 클러스터에 추가
    
map
```

```python
# 마커 위치에 미세한 변화를 주어 겹치지 X, 불규칙한 분포

import folium
import pandas as pd
import numpy as np

# 잠실역 중심 지도
map = folium.Map(location=[37.5131, 127.1006], zoom_start=10)

# 원을 그릴 위치의 위도와 경도를 설정
circle_lat_lon = [37.5131, 127.1006]

# folium.Circle() 메소드로 원 생성
folium.Circle(
    location=circle_lat_lon,
    radius=200,  # 원의 반지름 (미터 단위)
    color='crimson',
    fill=True,  # 테두리
    fill_color='crimson'
).add_to(map)

data = pd.read_csv("jamsilStation_cafe.csv")
marker_count = 0

# 데이터프레임에서 상가 업소의 위치를 지도에 표시
for index, row in data.iterrows():
    # 상가 업소의 위도와 경도에 미세한 변화 추가 
    location = (row['위도'] + np.random.uniform(-0.00005, 0.00005), row['경도'] + np.random.uniform(-0.00005, 0.00005))
    folium.Marker(location, icon = folium.Icon(icon = 'coffee', prefix = 'fa', color = 'red', icon_size = 35)).add_to(map)  # 개별 마커 추가
    marker_count += 1

map
```

```python
marker_count_df = pd.DataFrame({"Count": [marker_count]}, index = ['잠실역 200m 반경 내 카페'])
marker_count_df
```
